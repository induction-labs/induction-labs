```python
from __future__ import annotations

import pandas as pd

# Read into a flat DataFrame
df = pd.read_json("gs://induction-labs/action_capture/jeffrey/2025-06-16_101653_HTKN7/action_capture_000000.jsonl", lines=True)

# Normalize the nested 'action' dict into its own columns
df_flat = pd.json_normalize(df["action"])
df_final = pd.concat([df_flat, df["timestamp"]], axis=1)

df_final = df_final[(df_final["action"] == "mouse_button") | (df_final["action"] == "key_button")].reset_index(drop=True)
df_final
```
