```python
import pandas as pd
from PIL import Image
import os
base_path = "/home/ubuntu/.cache/huggingface/hub/datasets--OS-Copilot--ScreenSpot-v2/snapshots/5efbb1f1b5463a575f2eb7bc30fe29e49c15f93c/screenspotv2_image"

# Method 1: Using apply() - most pandas-like
def get_image_dimensions(img_path):
    img_path = os.path.join(base_path, img_path)
    try:
        with Image.open(img_path) as img:
            return pd.Series({'width': img.width, 'height': img.height})
    except Exception as e:
        print(f"Error loading {img_path}: {e}")
        return pd.Series({'width': None, 'height': None})

# Apply to DataFrame
```

```python
df[['width', 'height']] = df['img_filename'].apply(get_image_dimensions)
```
