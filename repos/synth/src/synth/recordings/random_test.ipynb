```python
import pandas as pd
base_path = "gs://induction-labs/passive_data/2025-08-14/reprocess_all_no_loading-04-30-15/".rstrip("/")
df = pd.read_json(f"{base_path}/samples.jsonl", lines=True)
df
```

```python
df['actions_with_thinking'] = df.apply(
    lambda r: list(zip(r['actions'], r['thinking'])),
    axis=1
)
df
```

```python
out = (
    df.assign(action_index=df['actions_with_thinking'].apply(lambda x: list(range(len(x)))))
      .explode(['actions_with_thinking', 'action_index'], ignore_index=False)   # use True if you want a fresh RangeIndex
)
out
```

```python
out['actions'] = out['actions_with_thinking'].apply(lambda x: x[0])
out['thinking'] = out['actions_with_thinking'].apply(lambda x: x[1])
out = out.drop(columns=['actions_with_thinking'])
out
```

```python
out.to_json(f"{base_path}/samples_actions_explode.jsonl", orient='records', lines=True)
```

```python
df_only_click = out[out['actions'].apply(lambda x: x.startswith(('click', 'left_double', 'right_single', 'drag')))]
df_only_click
```

```python
df_only_click.to_json(f"{base_path}/samples_actions_explode_only_click.jsonl", orient='records', lines=True)
```

```python
from smart_open import open as smart_open
import json
from synth.recordings.gemini_process import SaveAction


def load_item(source_dir: str, attempt_id: str, index: int) -> SaveAction:
  source_dir = source_dir.rstrip("/")
  with smart_open(f"{source_dir}/metadata/{attempt_id}.json", "r") as f:
    items = json.load(f)
    assert isinstance(items, list)
  return SaveAction.model_validate(items[index])
```

```python
from __future__ import annotations
from PIL import Image, ImageDraw


from io import BytesIO
from typing import IO
from PIL import Image
import base64
import requests

def load_png_from_path(path: str, mode: str | None = None) -> Image.Image:
    """Load a PNG from disk. Optionally convert to a mode like 'RGBA' or 'RGB'."""
    with Image.open(path) as im:
        return im.convert(mode) if mode else im.copy()  # copy() to detach from file
```

```python
import PIL
import PIL.Image
from synth.recordings.action_models import Point

def draw_red_circle(
    image: PIL.Image.Image, point: Point, radius: int = 15
) -> PIL.Image.Image:
    """
    Draw a highly visible red circle on the image at the specified point.
    """
    image = image.copy()
    # image = image.convert("L").convert("RGB")  # L = grayscale, then back to RGB for color drawing
    draw = ImageDraw.Draw(image)
    
    # Option 1: Larger, thicker circle with high contrast
    # draw.ellipse(
    #     [point.x - radius, point.y - radius, point.x + radius, point.y + radius],
    #     outline="red",
    #     width=4,
    #     fill=None
    # )
    
    # Option 2: Add a bright fill with contrasting outline
    draw.ellipse(
        [point.x - radius, point.y - radius, point.x + radius, point.y + radius],
        outline="red",
        # fill="red",
        width=10,
        # fill="green"  # or a semi-transparent red
    )
    
    # Option 3: Double circle for maximum visibility
    # draw.ellipse(
    #     [point.x - radius, point.y - radius, point.x + radius, point.y + radius],
    #     outline="red",
    #     width=4
    # )
    # draw.ellipse(
    #     [point.x - (radius-5), point.y - (radius-5), point.x + (radius-5), point.y + (radius-5)],
    #     outline="yellow",
    #     width=2
    # )
    
    return image
```
